Подготовка DF используя API Госзатраты

```python
import requests
import pandas as pd
```

```python
def getjson(url, data = None) :
    response = requests.get(url, params = data)
    print(response.url)
    response = response.json()
    return response
```

Берем данные из контрактов используя поиск

```python
url="http://openapi.clearspending.ru/restapi/v3/contracts/search/?"
search = {'pricerange':'10000-5000000'}
```

Синтаксис поискового запроса https://github.com/datacoon/clearspending-examples/wiki/%D0%9E%D0%BF%D0%B8%D1%81%D0%B0%D0%BD%D0%B8%D0%B5-API-%D0%9A%D0%BE%D0%BD%D1%82%D1%80%D0%B0%D0%BA%D1%82%D1%8B

```python
response = getjson(url, search)
```

```python
contracts = response['contracts']
data = contracts['data']
df = data[0]
df = pd.DataFrame(data)
df
```
